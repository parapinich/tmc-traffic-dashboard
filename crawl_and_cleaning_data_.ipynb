```python
import re

def clean_text(text):
  # Remove URLs
  text = re.sub(r'http\S+|www\S+|https\S+', '', text, flags=re.MULTILINE)
  # Remove time patterns (e.g., 20.45, 16.30 Wib, 17:00+07:00)
  text = re.sub(r'\d{1,2}\.\d{2}|\d{1,2}:\d{2}(\+\d{2}:\d{2})?(\s?wib)?', '', text, flags=re.IGNORECASE)
  # Remove "+numbers" (assuming they are like +0000 in timestamps)
  text = re.sub(r'\+\d+', '', text)
  # Remove specific characters like (), #, and .
  text = re.sub(r'[()#.]', '', text)
  # Convert to lowercase
  text = text.lower()
  return text

# Apply the cleaning function to the 'full_text' column
filtered_df['cleaned_text'] = filtered_df['full_text'].apply(clean_text)

# Select the relevant columns (you might want to keep the original 'full_text' or just use 'cleaned_text')
# For this task, let's save the 'created_at' and 'cleaned_text' columns
cleaned_data_df = filtered_df[['created_at', 'cleaned_text']]

# Save the cleaned data to a new CSV file
cleaned_data_df.to_csv("clean_data.csv", index=False)

print("\nCleaned data saved to 'clean_data.csv'. Head of the cleaned data:")
print(cleaned_data_df.head())
```

### **POS TAGGING**

```python
import stanza
import pandas as pd


# Initialize Stanza for Indonesian with the 'pos' processor
try:
    nlp = stanza.Pipeline(lang='id', processors='tokenize,pos')
except Exception as e:
     print(f"Error initializing Stanza pipeline: {e}")
     nlp = None

def stanza_pos_tagging(text):
    if nlp:
        doc = nlp(text)
        pos_tags = []
        for sentence in doc.sentences:
            for word in sentence.words:
                pos_tags.append((word.text, word.upos)) # word.upos gives Universal POS tags
        return pos_tags
    else:
        return []


# Assuming 'cleaned_data_df' is your DataFrame with the 'cleaned_text' column
# Apply the Stanza POS tagging function to the 'cleaned_text' column
if nlp:
    cleaned_data_df['stanza_pos_tags'] = cleaned_data_df['cleaned_text'].apply(stanza_pos_tagging)

    # Display the DataFrame with Stanza POS tags
    print("\nDataFrame with Stanza POS tags:")
    display(cleaned_data_df.head())

    # Optionally, save the DataFrame with Stanza POS tags
    cleaned_data_df.to_csv("clean_data_with_stanza_pos.csv", index=False)
    print("\nData with Stanza POS tags saved to 'clean_data_with_stanza_pos.csv'.")
else:
    print("Stanza pipeline could not be initialized. Skipping POS tagging.")
```

### **BIO TAGGING**

```python
import pandas as pd
import json

# 1. Memuat dataset
try:
    # Penting: Pastikan CSV Anda memiliki kolom 'created_at'
    df = pd.read_csv('clean_data_with_stanza_pos.csv')
    print("✅ Dataset 'clean_data_with_stanza_pos.csv' berhasil dimuat.")
except FileNotFoundError:
    print("❌ GAGAL: File 'clean_data_with_stanza_pos.csv' tidak ditemukan.")
    df = pd.DataFrame()

# ... (Fungsi generate_bio_tags_v3 dan KEYWORDS tetap sama persis seperti sebelumnya) ...
KEYWORDS = {
    "FROM": {"di"}, "TO": {"arah", "mengarah", "untuk"}, "TO_2_TRIGGER": {"maupun", "sebaliknya", "dan"},
    "STATUS": {"terpantau"}, "OBSTACLE": {"kecelakaan", "banjir", "longsor"},
    "OBSTACLE_MULTIWORD": {"jalan rusak"}, "WEATHER": {"hujan", "cerah", "mendung"}
}
GREEDY_STOP_WORDS = KEYWORDS["TO"] | KEYWORDS["STATUS"] | KEYWORDS["TO_2_TRIGGER"]
def generate_bio_tags_v3(text):
    tokens = text.lower().split()
    bio_tags = ['O'] * len(tokens)
    to2_triggered = False
    i = 0
    while i < len(tokens):
        token = tokens[i]
        if token in KEYWORDS["TO_2_TRIGGER"]: to2_triggered = True
        if token in KEYWORDS["WEATHER"]: bio_tags[i] = 'B-WEATHER'
        if token in KEYWORDS["OBSTACLE"]: bio_tags[i] = 'B-OBSTACLE'
        if ' '.join(tokens[i:i+2]) == 'jalan rusak':
            bio_tags[i] = 'B-OBSTACLE'
            if i + 1 < len(tokens): bio_tags[i+1] = 'I-OBSTACLE'
        if token in KEYWORDS["FROM"]:
            j = i + 1
            if j < len(tokens) and tokens[j] not in GREEDY_STOP_WORDS:
                bio_tags[j] = 'B-FROM'
                j += 1
            while j < len(tokens) and tokens[j] not in GREEDY_STOP_WORDS:
                bio_tags[j] = 'I-FROM'
                j += 1
            i = j - 1
        elif token in KEYWORDS["TO"]:
            if i + 1 < len(tokens):
                if to2_triggered:
                    bio_tags[i+1] = 'B-TO2'; to2_triggered = False
                else: bio_tags[i+1] = 'B-TO'
        elif token in KEYWORDS["STATUS"]:
            if i + 1 < len(tokens): bio_tags[i+1] = 'B-STATUS'
            if i + 2 < len(tokens) and tokens[i+2] not in GREEDY_STOP_WORDS:
                bio_tags[i+2] = 'I-STATUS'
        i += 1
    return tokens, bio_tags


# 3. Proses Ekstraksi dan Pembuatan Label
labeled_results = []
if not df.empty and 'created_at' in df.columns:
    for index, row in df.iterrows():
        text = str(row['cleaned_text'])
        # --- PERUBAHAN DI SINI: Ambil juga data 'created_at' ---
        created_at_val = str(row['created_at'])

        tokens, bio_tags = generate_bio_tags_v3(text)

        sentence_labels = []
        for i in range(len(tokens)):
            sentence_labels.append({"token": tokens[i], "bio_tag": bio_tags[i]})

        # --- PERUBAHAN DI SINI: Sertakan 'created_at' dalam hasil ---
        labeled_results.append({
            "text": text,
            "created_at": created_at_val, # Tambahkan baris ini
            "labeled_tokens": sentence_labels
        })
else:
    if df.empty:
        pass
    else:
        print("❌ Peringatan: Kolom 'created_at' tidak ditemukan di file CSV.")


# 4. Simpan Hasil ke File JSON
output_filename = 'hasil_bio_labeling_final.json'
with open(output_filename, 'w', encoding='utf-8') as f:
    json.dump(labeled_results, f, indent=4, ensure_ascii=False)
```

### **EKTRAKSI**

```python
import json
import pandas as pd

def convert_bio_to_split_entities(labeled_data_path, output_path):

    try:
        with open(labeled_data_path, 'r', encoding='utf-8') as f:
            data = json.load(f)
        print(f"✅ Berhasil memuat file '{labeled_data_path}'.")
    except FileNotFoundError:
        print(f"❌ GAGAL: File '{labeled_data_path}' tidak ditemukan.")
        return

    final_results = []

    for sentence_data in data:
        text = sentence_data['text']
        labeled_tokens = sentence_data['labeled_tokens']
        # Ambil created_at dari data
        created_at_str = sentence_data.get('created_at')

        # Ekstrak semua entitas seperti biasa
        entities = {}
        i = 0
        while i < len(labeled_tokens):
            tag = labeled_tokens[i]['bio_tag']
            if tag.startswith('B-'):
                entity_type = tag.split('-')[1]
                entity_tokens = [labeled_tokens[i]['token']]
                j = i + 1
                while j < len(labeled_tokens) and labeled_tokens[j]['bio_tag'] == f'I-{entity_type}':
                    entity_tokens.append(labeled_tokens[j]['token'])
                    j += 1
                if entity_type == 'OBSTACLE':
                    entities.setdefault(entity_type, []).append(' '.join(entity_tokens))
                else:
                    entities[entity_type] = ' '.join(entity_tokens)
                i = j - 1
            i += 1

        if created_at_str:
            try:
                # Konversi string ke objek datetime menggunakan pandas
                dt_object = pd.to_datetime(created_at_str)
                # Format menjadi DD-MM-YYYY dan HH:MM
                entities['DATE'] = dt_object.strftime('%d-%m-%Y')
                entities['TIME'] = dt_object.strftime('%H:%M')
            except (ValueError, TypeError):
                # Jika formatnya salah, biarkan kosong
                entities['DATE'] = None
                entities['TIME'] = None

        # Logika untuk memecah data jika ada TO2
        if 'TO2' in entities:
            entry1 = entities.copy()
            entry1.pop('TO2')
            final_results.append({"text": text, "entities": entry1})

            entry2 = entities.copy()
            entry2['TO'] = entry2.pop('TO2')
            final_results.append({"text": text, "entities": entry2})
        elif entities:
            final_results.append({"text": text, "entities": entities})

    # Simpan hasil akhir
    if final_results:
        with open(output_path, 'w', encoding='utf-8') as f:
            json.dump(final_results, f, indent=4, ensure_ascii=False)
        print(f"\n✅ Konversi selesai! Total {len(final_results)} data dihasilkan.")
        print(f"Hasil disimpan di file '{output_path}'.")

# --- Jalankan Konversi ---
input_bio_file = 'hasil_bio_labeling_final.json'
output_extraction_file = 'hasil_ekstraksi_final_split.json'

convert_bio_to_split_entities(input_bio_file, output_extraction_file)
```
